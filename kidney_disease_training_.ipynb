```python
import pandas as pd
```

```python
dt=pd.read_csv('/content/kidney_disease.csv')
dt.head()
```

```python
dt.dropna(inplace=True)
```

```python
dt.info()
```

```python
dt['rbc'].unique()
```

```python
for i in dt.columns:
  if dt[i].dtype=='object':
    dt[i]=dt[i].astype('category').cat.codes
  dt.info()
```

```python
import keras
```

```python
from keras.models import Sequential
model=Sequential()
```

```python
from keras.layers import Dense
```

```python
dt.head()
```

```python
dt.corr()
```

```python
dt=dt.drop(['su','ba','wc','rc','cad'],axis=1)
```

```python
dt.head()
```

```python
x=dt.iloc[:,0:-1]
y=dt['classification']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
print(x.shape,y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.build(input_shape=(None,20))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score  #Evaluation
accuracy_score(y_test,y_pred)
```

```python
model.save('kidney.h5')
```
